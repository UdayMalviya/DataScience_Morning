```python
# list[dict]
data = [
{"name": 'ram', 'age': 30, 'sal': '50000'},
{"name": 'shyam', 'age': 25, 'sal': '70000'},
{"name": 'ravi', 'age': 28, 'sal': '40000'}]
```

```python
data
```

```python
df = pd.DataFrame(data, index=pd.RangeIndex(1001, 1004, name='index'))
```

```python
df
```

```python
df.info()
```

```python
df['sal'] = df['sal'].astype('int')
```

```python
df.reset_index(drop=False)
```
