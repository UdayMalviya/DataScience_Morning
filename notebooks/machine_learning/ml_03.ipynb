```python
# OrdinalEncoding 
# nominal, ordinal

X = pd.DataFrame(['low', 'medium', 'high', 'medium', 'low', 'low'], columns=['categories'])
```

```python
X
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encode = OrdinalEncoder(categories=[['low', 'medium', 'high']])
```

```python
encode.fit_transform(X)
```

```python
encode.categories_
```

```python
encode.transform([['low'], ['medium']])
```
