```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x)
```

```python
x.shape, y.shape
```

```python
fig, axs =  plt.subplots(1, figsize=(5,2))
axs.plot(x, y, '-g', label='sin')
axs.plot(x, y*2, '--r', label='sin*2')
axs.set_title("sin wave")
axs.set_xlabel('X axis')
axs.set_ylabel("Y axis")
axs.legend(loc='lower left')

plt.show()
```

```python
# histogram
np.random.seed(42)
numbers = np.random.randn(100)
```

```python
numbers
```

```python
fig, ax = plt.subplots(1)
ax.hist(numbers, bins=15, color='red', alpha=0.5)
ax.hist(numbers*2, bins=15, color='blue', alpha=0.2)
plt.grid()
plt.show()
```

```python
# bar  
import pandas as pd
df = pd.DataFrame({"labels": ['A', "B", 'C'], 'values':[100, 500, 800], 'width':[0.1, 0.3, 0.2]})
```

```python
df
```

```python
plt.figure(figsize=(5,4))
plt.bar(df['labels'], df['values'], width=df['width'], color=['red', 'green', 'yellow'], label='plot1')
plt.bar(df['labels'], df['values']*10, width=df['width'], alpha=0.2, label='plot2')
plt.title("title")
plt.legend()
plt.show()
```
