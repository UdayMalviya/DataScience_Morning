```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplot_mosaic([['A', 'B']])
ax['A'].plot(np.array([1,3]), color=(0.9, 0.8, 0.6))
ax['B'].scatter(np.array([5,6]), np.array([10, 12]), cmap='Set2', c=[3,4])
plt.show()
```

```python
x = np.linspace(1, 100, 100)
y = np.log(x)
c = np.random.randint(0, 255, 100)
```

```python
plt.scatter(x, y, c=c, cmap='Pastel1')

plt.text(20, 3, 'Text', size=14.0, c='r')
plt.axline((0,2), slope=0.5)
plt.axvline(20)
plt.axhline(20)
```
