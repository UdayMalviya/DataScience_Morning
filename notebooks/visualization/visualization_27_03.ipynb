```python
sns.histplot(data['total_bill'], bins=15, kde=True) # more memory
```

```python
sns.displot(x=data['tip'], y= data['total_bill'],  kind='hist', color='r')
```

```python
sns.boxplot(x=data['day'], y = data['tip'], hue=data['sex'])
sns.boxplot(data)
```
