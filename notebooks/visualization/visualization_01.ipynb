```python
a = np.linspace(0, 2 * np.pi, 1000) # 0 - 2pi
```

```python
b = np.sin(a)
c = np.cos(a)
t = np.tan(a)
```

```python
plt.plot(a, b)
plt.plot(a, c)
plt.plot(a, t)
plt.show()
```

```python
fig, axs = plt.subplots(1, 3)
axs[0].plot(b, a) # sin
axs[1].plot(c, a) # cos
axs[2].plot(t, a)
# plt.show() np.int8
```

```python
# 1/ sin() # cosec
# 1/cos() # sec
# 1/tan() # cot
```
