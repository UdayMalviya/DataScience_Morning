```python
# min and max 
# df['Store'].agg(['min', 'max'])
def find_min_max(data: pd.DataFrame, col: str):
    """Returns the minimum and maximum of a series"""
    try:
        if col not in data.columns:
            raise KeyError(f" column is not present in the dataframe")
        minimum = data[col].min()
        maximum = data[col].max()
        return minimum, maximum
    except Exception as e:
        print(f"Got error while finding min_max: {e}")
```

```python
# convert_dtypes
def convert_dtypes(df: pd.DataFrame, col:str,  dtype = None, inplace: bool=False):
    try:
        if not dtype:
            print("Enter datatype and retry..")
        if inplace:
            df[col] = df[col].astype(dtype)
            return None
        else:
            series = df[col].astype(dtype)
            return series
    except Exception as e:
        print(f"Got error while converting dtypes: {e}")
    
```
